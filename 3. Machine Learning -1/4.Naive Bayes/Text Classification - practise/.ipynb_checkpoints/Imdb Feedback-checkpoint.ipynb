## SMS Spam Classifier: Multinomial Naive Bayes

### 1. Importing and Preprocessing Data

```python
import numpy as np
```

```python
import pandas as pd
# reading the training data
docs = pd.read_csv('movie_review_train.csv')
docs_test = pd.read_csv('movie_review_test.csv')
docs.head()
```

```python
# number of SMSes / documents
len(docs)
```

```python
# counting spam and ham instances
ham_spam = docs["class"].value_counts()
ham_spam
```

```python
print("spam rate is about {0}%".format(
    round((ham_spam[1]/float(ham_spam[0]+ham_spam[1]))*100), 2))
```

```python
# mapping labels to 0 and 1
docs['class'] = docs["class"].map({'Neg':0, 'Pos':1})
docs_test['class'] = docs_test["class"].map({'Neg':0, 'Pos':1})
```

```python
docs_test.head()
```

```python
# convert to X and y
X_train = docs["text"]
y_train = docs["class"]
print(X_train.shape)
print(y_train.shape)
```

```python
# convert to X and y
X_test = docs_test["text"]
y_test = docs_test["class"]
print(X_test.shape)
print(y_test.shape)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
X_test.head()
y_test.head()
```

```python
# vectorizing the sentences; removing stop words
from sklearn.feature_extraction.text import CountVectorizer
vect = CountVectorizer(stop_words='english')
```

```python
vect.fit(X_train)
```

```python
# printing the vocabulary
vect.vocabulary_
```

```python
# vocab size
len(vect.vocabulary_.keys())
X_train_vocabs = vect.get_feature_names()
len(X_train_vocabs)
```

```python
vect3 = CountVectorizer(stop_words='english',min_df=.03,max_df=.8)
vect3.fit(X_train)
X_train_vocabs = vect3.get_feature_names()
len(X_train_vocabs)
```

```python
# transforming the train and test datasets
X_train_transformed = vect3.transform(X_train)
X_test_transformed = vect3.transform(X_test)
```

```python
# note that the type is transformed (sparse) matrix
print(type(X_train_transformed))
print(X_train_transformed)
```

```python
X_train_transformed.toarray()
```

```python
pd.DataFrame(X_train_transformed.toarray(), 
             columns=vect3.get_feature_names())
```

```python
print(type(X_test_transformed))
print(X_test_transformed)
```

```python
X_train_transformed.toarray()
```

```python
pd.DataFrame(X_train_transformed.toarray(), 
             columns=vect3.get_feature_names())
```

### 2. Building the Model: Multinomial Naive Bayes

```python
# training the NB model and making predictions
from sklearn.naive_bayes import MultinomialNB
mnb = MultinomialNB()
# fit
mnb.fit(X_train_transformed,y_train)

# predict class
y_pred_class = mnb.predict(X_test_transformed)

# predict probabilities
y_pred_proba = mnb.predict_proba(X_test_transformed)
```

```python
# note that alpha=1 is used by default for smoothing
mnb
```

```python
y_pred_class
```

```python
# probability of each class (test data)
print("probability of test document belonging to class Positive" , y_pred_proba[:,0])
print("probability of test document belonging to class Negative" , y_pred_proba[:,1])
```

### 3. Building the Model: Bernoulli Naive Bayes

```python
from sklearn.naive_bayes import BernoulliNB

# instantiating bernoulli NB class
bnb=BernoulliNB()

# fitting the model
bnb.fit(X_train_transformed, y_train)

y_pred_bnb_class = mnb.predict(X_test_transformed)

# also works
# bnb.fit(X_transformed.toarray(), y_train)

# predicting probability of test data

y_pred_prob_bnb = bnb.predict_proba(X_test_transformed)
y_pred_prob_bnb
```

```python
# note that alpha=1 is used by default for smoothing
bnb
```

```python
y_pred_bnb_class
```

```python
# probability of each class (test data)
print("probability of test document belonging to class Positive" , y_pred_prob_bnb[:,0])
print("probability of test document belonging to class Negative" , y_pred_prob_bnb[:,1])
```

### Model Evaluation for Multinomial NB

```python
# printing the overall accuracy
from sklearn import metrics
metrics.accuracy_score(y_test, y_pred_class)
```

```python
# confusion matrix
metrics.confusion_matrix(y_test, y_pred_class)
# help(metrics.confusion_matrix)
```

```python
confusion = metrics.confusion_matrix(y_test, y_pred_class)
print(confusion)
TN = confusion[0, 0]
FP = confusion[0, 1]
FN = confusion[1, 0]
TP = confusion[1, 1]
```

```python
sensitivity = TP / float(FN + TP)
print("sensitivity",sensitivity)
```

```python
specificity = TN / float(TN + FP)
print("specificity",specificity)
```

```python
precision = TP / float(TP + FP)
print("precision",precision)
print(metrics.precision_score(y_test, y_pred_class))
```

```python
print("precision",precision)
print("PRECISION SCORE :",metrics.precision_score(y_test, y_pred_class))
print("RECALL SCORE :", metrics.recall_score(y_test, y_pred_class))
print("F1 SCORE :",metrics.f1_score(y_test, y_pred_class))
```
