#### Intro to ML

Machine learning models can be classified into the following three types based on the task performed and the nature of the output:

1. Regression: The output variable to be predicted is a continuous variable, e.g. scores of a student

2. Classification: The output variable to be predicted is a categorical variable, e.g. classifying incoming emails as spam or ham

3. Clustering: No pre-defined notion of a label is allocated to groups/clusters formed, e.g. customer segmentation 

Machine learning models into two broad categories:

1. Supervised learning methods : Past data with labels is used for building the model. Regression and classification algorithms fall under this category

2. Unsupervised learning methods: No pre-defined labels are assigned to past data.Clustering algorithms fall under this category



#### Simple Linear Regression

A simple linear regression model attempts to explain the relationship between a dependent and an independent variable using a straight line.Depedent variables have to be continous values

The independent variable is also known as the predictor variable. And the dependent variables are also known as the output variables.

$ y = {\beta}_{0} + {\beta}_{1}X $

where ${\beta}_{1}$ is a slope and ${\beta}_{0}$ is intercept

$\hat{Y} = \hat{\beta}_{0} + \sum \limits _{j=1} ^{p} X_{j}\hat{\beta}_{j} $

#### Best- Fit Line

The best will be the one which will pass through maximum points. One way of finding such line is to Residual Sum of Squares (RSS) which is square of distance b/w points and predicated line points. 

$ {e}_{i} = y_{i} - {y}_{pred} $

$ RSS = {e^{2}}_{1} + {e^{2}}_{2} + ..... + {e^{2}}_{n}  $ .......(1)

The best fit line will have minimum RSS value.

Since $ y_{pred} $ will fall on the best fit line. It can be calcuated via 

$ {y}_{pred} = {\beta}_{0} + {\beta}_{1}X_{i} $ ...... (2)


substituting eq(1) in eq(2)..

$ RSS = (y_{1} - {\beta}_{0} - {\beta}_{1}X_{1})^{2} + (y_{2} - {\beta}_{0} - {\beta}_{1}X_{2})^{2} + (y_{3} - {\beta}_{0} - {\beta}_{1}X_{3})^{2} + ...... (y_{n} - {\beta}_{0} - {\beta}_{1}X_{n})^{2}$

$ RSS = \sum \limits _{i=1} ^{n} (y_{i} - {\beta}_{0} - {\beta}_{1}X_{i})^{2}   $

$ Cost Function = \sum \limits _{i=1} ^{n} (y_{i} - {\beta}_{0} - {\beta}_{1}X_{i})^{2}   $
#### Optimisation 

There are two types of optimisation: constrained and unconstrained optimisation. Optimisation can be of 2 types minimisation and maximaization. 

In most cases, you will have to minimise the cost function. Minimising and maximising a function is very simple:

Differentiate the function w.r.t the parameter and equate to 0.

For minimisation  the function value of the double differential should be greater than 0.

For maximisation - the function value of the double differential should be less than 0.


1. Constrainted Minimisation - Differentiate the function w.r.t the parameter and equate to non-zero value. i.e. a condition on $ {\beta}_{0} and {\beta}_{1} $ such as  $ {\beta}_{0} > 2  and  {\beta}_{1} < -1 $ etc.

2. Unconstrained Minimisation - Differentiate the function w.r.t the parameter and equate to 0.


Unconstrained minimisation can be solved using two methods:

1. <b> Closed form method: </b> The function to be minimised is simply differentiated and equated to 0 to achieve a solution. The solution is also double differentiated to check if the solution is greater than 0.Cost function varies according to the algorithm being used. In this segment, for fitting a straight line the cost function was the sum of squared errors, but it will vary from algorithm to algorithm. To minimise the sum of squared errors and find the optimal $ {\beta}_{0} and {\beta}_{1} $, we differentiated the sum of squared errors w.r.t the parameters  $ {\beta}_{0} and {\beta}_{1} $. We then solved the linear equations to obtain the values of $ {\beta}_{0} and {\beta}_{1} $. 


2. Itertive Form Method: There are 2 types of Iterative forms:

    i. <b> First Order or Gradient Descent:</b> It is an iterative minimisation method which reaches the minima step by step .You start with an initial assumed value of the parameter. This initial assumed value can be anything (say $ X_{0} $). Then you assume $ \alpha $ which is rate of learning. For that value $ X_{0} $ , you calculate the output of the differentiated function which we denote as $ f^{'}(x) $. Then the new value of the parameter becomes $ x -f^{'}(x)*\alpha $ . You continue the process untill the algorithm reaches an optimum point ($ X_{4} $); i.e the value of the parameter does not change effectively after this point. RSS: Residual Sum of Squares
TSS: Total Sum of Squares

$ TSS = \sum \limits _{i=1} ^{n} (y_{i} - \bar{y})^{2}   $

The strength of a linear regression model is mainly explained by R,  where R = 1 - (RSS / TSS)

$ RSE = \sqrt RSS/(N-2)  $

### Assumption for Linear Regression

1. Linear Relation b/x X and Y: X and Y should display some sort of a linear relationship; otherwise, there is no use of fitting a linear model between them.

2. Error Terms are normally distributed with mean 0 
3. Error Terms are independent of each other
4. Error Term has constant variance (homoscedasticity)

No distribution on distribution of X and Y

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import sklearn
from sklearn.model_selection import  train_test_split
from sklearn.metrics import mean_squared_error 
from sklearn.metrics import r2_score
```

```python

## Read data
advertising = pd.read_csv("advertising.csv")
advertising.head()
```

```python
advertising.describe()
X=advertising["TV"]
y=advertising["Sales"]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.7,test_size=0.3,random_state=100)
```

```python
X_train_sm=sm.add_constant(X_train)
X_train_sm.head(5)
y_train.head(5)
```

```python
lr_model=sm.OLS(y_train, X_train_sm).fit()
lr_model.params
### Sales = 6.948683 + 0.054546TV
```

```python
lr_model.summary()
```

```python
plt.scatter(X_train,y_train)
plt.plot(X_train,6.948683 + 0.054546*X_train,'r')
plt.show()
```

```python
y_train_pred= lr_model.predict(X_train_sm)
res=y_train - y_train_pred
sns.distplot(res,bins=15)
plt.title('Error Terms', fontsize=15)
plt.xlabel('y_train - y_pred',fontsize=15)
plt.show()
```

```python
plt.scatter(X_train,res)
```

```python
X_test_sm=sm.add_constant(X_test)
y_pred = lr_model.predict(X_test_sm)
```

```python
y_pred.head(5)
```

```python
np.sqrt(mean_squared_error(y_pred, y_test))
```

```python
r_squared=r2_score(y_test,y_pred)
r_squared
```
