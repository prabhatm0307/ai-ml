```python
class Solution:
    def twoSum(self,nums: list[int],target: int) -> list[int]:
        seen={}
        for i,v in enumerate(nums):
            remaining=target-v
            if remaining in seen:
                print([seen[remaining],i])
                return [seen[remaining],i]
            seen[v]=i
        return []
if __name__ == '__main__':
    s=Solution()
    s.twoSum([2,7,5,3],9)
    s.twoSum([3,2,4],6)
## time complexity=O(n)
```

```python
### sliding window 
class Solution:
    def LongestStringWithoutRepeatingCharacters(self,s)-> int:
        charSet=set()
        i=0
        res=0
        for x in range(len(s)):
            while s[x] in charSet:
                charSet.remove(s[i])
                i=i+1
            charSet.add(s[x])
            res=max(res,x-i+1)
        print(res)
        return res
if __name__ == '__main__':
    s=Solution()
    s.LongestStringWithoutRepeatingCharacters("pwwke")
# time complexity : O(n)
# space complexity: O(n)
```

```python
'''
Approach:
1. sort the array
2. start first loop to scan thru all items 
3. pass if next is same
4. assign left as next item and right as last item of the list
5. loop till left < right, calculate sum and check if sum <0 then shift left else if sum>0 shift right
    else right list is found,append it to result list,
'''
class Solution:
    def threeSum(self,nums:list[int])-> list[list[int]]:
        res=[]
        nums.sort()
        for i,v in enumerate(nums):
            if(i>0 and nums[i-1]==v):
                continue
            l,r=i+1,len(nums)-1
            while l<r:
                thrsum=v+nums[l]+nums[r]
                if thrsum>0:
                    r=r-1
                elif thrsum<0:
                    l+=1
                else:
                    res.append([v,nums[l],nums[r]])
                    l=l+1
                    while nums[l]==nums[l-1] and l<r:
                        l=l+1
        return res
if __name__ == '__main__':
    s=Solution()
    res=s.threeSum([-1,0,1,2,-1,-4])
    print(res)
# time complexity : O(n^2)
# space complexity: O(n)
```

```python
class Solution:
    def threeSumClosest(self, nums: list[int], target: int) -> int:
        nums.sort()
        res=sum(nums[:3])
        for i in range(len(nums)):
            l,r=i+1,len(nums)-1
            while l<r:
                sumhere=nums[i]+nums[l]+nums[r]
                if abs(res-target)> abs(sumhere-target):
                    res=sumhere
                if sumhere < target:
                    l=l+1
                else:
                    r=r-1
        return res
if __name__ == '__main__':
    s=Solution()
    res=s.threeSumClosest([-1,0,1,2,-1,-4],1)
    print(res)
# time complexity : O(n^2)
# space complexity: O(n)                    
```

```python
class Solution:
    def coinChange(self, coins: list[int], amount: int) -> int:
        dp = [float('inf')] * (amount + 1)
        dp[0] = 0        
        for coin in coins:
            for x in range(coin, amount + 1):
                dp[x] = min(dp[x], dp[x - coin] + 1)
        return dp[amount] if dp[amount] != float('inf') else -1
if __name__ == '__main__':
    s=Solution()
    res=s.coinChange([1,3,4,5],11)
    print(res)      
```

```python
class Solution:
    def bestTime(self, nums:list[int]) -> int:
        profit=0
        for i,v in enumerate(nums):
            l,r=i+1,len(nums)
            while l<r:
                if v>nums[l]:
                    l+=1
                    continue
                else:
                    profit=max(profit,nums[l]-v)
                    l+=1
        return profit
    
if __name__ == '__main__':
    s=Solution()
    res=s.bestTime([7,1,5,3,6,4])
    print(res)
```

```python
class Solution:
    def maxProfit(self, prices: list[int]) -> int:
        minprice=100000
        profit=0
        for i,v in enumerate(prices):
            if(v<minprice):
                minprice=v
            elif(v-minprice>profit):
                profit=v-minprice
        return profit
if __name__ == '__main__':
    s=Solution()
    res=s.maxProfit([7,1,5,3,6,4])
    print(res)
```

```python
class Solution:
    def maxProfit(self, prices: List[int]) -> int:
        profit=0
        for i,v in enumerate(prices):
            l,r=i+1,len(prices)
            while l<r:
                if v>prices[l]:
                    l+=1
                    continue
                else:
                    profit=max(profit,prices[l]-v)
                    l+=1
        return profit
```

```python
class Solution:
    def longestPalindrome(self, s:str)-> str:
        self.maxlen = 0
        self.start = 0

        for i in range(len(s)):
            self.expandFromCenter(s,i,i)
            self.expandFromCenter(s,i,i+1)
        return s[self.start:self.start+self.maxlen]


    def expandFromCenter(self,s,l,r):
        while l > -1 and r < len(s) and s[l] ==s[r]:
            l -= 1
            r += 1

        if self.maxlen < r-l-1:
            self.maxlen = r-l-1
            self.start = l+1
if __name__ == '__main__':
    s=Solution()
    res=s.longestPalindrome('cabac')
    print(res)  
## Time Comp = O(n2)
## Space COmp = O(1)
```

```python
class Solution:
    def Xylopyric(self,n:int)-> str:
        t=n*n
        s=str(t)
        flg=False
        if(len(s)%2==0):
            mid=int(len(s)/2)
            if(int(s[:mid])+int(s[mid:])==n  and (int(s[mid:])!= 0 and int(s[:mid])!=0)):
                flg=True
        if(len(s)%2!=0):
            mid=int(len(s)/2)
            mid_v=[mid,mid+1]
            for i in mid_v:
                if(int(s[:i])+int(s[i:])==n and int(s[mid:])!=0 and int(s[:mid]!=0)):
                    flg=True
                    break
        if flg:
            return 'Yes'
        else:
            return 'No'
        
if __name__ == '__main__':
    s=Solution()
    res=s.Xylopyric(297)
    print(res)          
```

```python
class Solution:
    def Xylopyric(self,n:int)-> str:
        t=n*n
        s=str(t)
        mid=int(len(s)/2)
        flg=self.validNum(s,n,mid)
        if(len(s)%2!=0):
            flg=self.validNum(s,n,mid+1)
        if flg:
            return 'Yes'
        else:
            return 'No'
    def validNum(self,s,n,mid):
        if(int(s[:mid])+int(s[mid:])==n  and (int(s[mid:])!= 0 and int(s[:mid])!=0)):
            return True
            
        else:
            return False
if __name__ == '__main__':
    s=Solution()
    res=s.Xylopyric(297)
    print(res)    
```
